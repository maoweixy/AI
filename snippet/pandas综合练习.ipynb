# 2002 年-2018 年上海机动车拍照拍卖

```python
df = pd.read_csv('数据集/2002年-2018年上海机动车拍照拍卖.csv')
df.head().append(df.tail())
```

## 哪一次拍卖的中标率首次小于 5%?

```python
df['bidding_rate'] = df['Total number of license issued'] / df['Total number of applicants']
df[df['bidding_rate']<0.05].head(1)
```

## 将第一列时间列拆分成两个列，一列为年份(格式为 20××)，另一列为 月份(英语缩写)，添加到列表作为第一第二列，并将原表第一列删除， 其他列依次向后顺延

```python
year = []
month = []
for f in list(df['Date'].values):
    year.append(f.split('-')[0])
    month.append(f.split('-')[1])
df['Year'] = year
df['Year'] = df['Year'].apply(lambda x: '200'+str(x) if len(str(x))==1 else '20'+str(x))
df['Month'] = month

df = df.reindex(columns=['Year', 'Month', 'Date', 'Total number of license issued', 'lowest price ', 'avg price', 'Total number of applicants', 'bidding_rate'])
df.drop(columns='Date', inplace=True)
df
```

## 按年统计拍卖最低价的下列统计量:最大值、均值、0.75 分位数，要求显示在同一张表上

```python
df.groupby('Year').describe(percentiles=[.75])
```

## 一般而言某个月最低价与上月最低价的差额，会与该月均值与上月均值的差额具有相同的正负号，哪些拍卖时间不具有这个特点?

```python
for index in range(202):
    flag = (df.loc[index,'lowest price ']-df.loc[index+1,'lowest price '])*\
    (df.loc[index,'avg price']-df.loc[index+1,'avg price'])
    if flag < 0:
        df.loc[index+1]
```

## 将某一个月牌照发行量与其前两个月发行量均值的差额定义为发行增 益，最初的两个月用 0 填充，求发行增益极值出现的时间

```python
df['issue gain'] = 0
for index in range(3,202):
    df.loc[index,'issue gain']= df.loc[index,'Total number of license issued']-(df.loc[index-1,'Total number of license issued']+df.loc[index-2,'Total number of license issued'])/2
    
df[df['issue gain']==df['issue gain'].min()]
df[df['issue gain']==df['issue gain'].max()]
```

# 2007 年-2019 年俄罗斯机场货运航班运载量问题

```python
df = pd.read_csv('数据集/2007年-2019年俄罗斯货运航班运载量.csv')
df.head().append(df.tail())
```

## 求每年货运航班总运量

```python
df.groupby('Year')['Whole year'].agg('sum')
```

## 每年记录的机场都是相同的吗?

```python
df.groupby('Year')['Airport name'].count()
```

## 按年计算 2010 年-2015 年全年货运量记录为 0 的机场航班比例

```python
# df_1 = [df['Year'] > 2009 and df['Year'] < 2016]
df_1 = df.query('Year > 2009 & Year < 2016')
df_1[df_1['Whole year'] == 0].shape[0]/df_1.shape[0]
```

## 若某机场至少存在 5 年或以上满足所有月运量记录都为 0，则将其所有年份的记录信息从表中删除，并返回处理后的表格

```python
airport_name = []
for name,group in df.groupby('Airport name'):
    zore_whole_year = group[group['Whole year']==0].shape[0]
    if zore_whole_year >= 5:
        airport_name.append(name)
airport_name
df = df.loc[~df['Airport name'].isin(airport_name)]
df
```

```python
df[df['Airport name'].isin(airport_name)]
```

## 在统计学中常常用秩代表排名，现在规定某个机场某年某个月的秩为该 机场该月在当年所有月份中货运量的排名(例如 *** 机场 19 年 1 月运 量在整个 19 年 12 个月中排名第一，则秩为 1)，那么判断某月运量情 况的相对大小的秩方法为将所有机场在该月的秩排名相加，并将这个量 定义为每一个月的秩综合指数，请根据上述定义计算 2016 年 12 个月 的秩综合指数。

```python
df = df[df['Year'] == 2016]
month = list(df.columns[2:14])
for i in month:
    df[i+'_r']=0
for index in df.index:
    mid = df.loc[index,month].sort_values(ascending = False)
    rank = 1
    for ii in mid.index:
        df.loc[index,ii+'_r']=rank
        rank+=1
print(df.iloc[:,-12:].sum())
```
